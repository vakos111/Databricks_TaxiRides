### A. Define File Path

```python
# COMMAND ----------

# DBTITLE 1,Define file path as variable
# Replace the values: ***ContainerName***, ***DataLakeName*** and file path (if required)

cabsFilePath = "abfss://de-storage-databrick@databrickstorage1.dfs.core.windows.net/Raw/Cabs.csv"

print(cabsFilePath)
```

### B. Read File by Inferring Schema


```python
# DBTITLE 1,Read data by inferring schema
# Create DataFrame by applying the schema

cabsDF = (
    spark
    .read
    .option("header", "true")

    .option("inferSchema", "true")

    .csv(cabsFilePath)
)

# Print schema
cabsDF.printSchema()

```

### C. Save Data to Data Lake as Spark (Delta) Table


```python
(
    cabsDF
    .write

    .mode("overwrite")

    .format("delta")

    .saveAsTable("taxicatalog.rides.cabs")
)
```
