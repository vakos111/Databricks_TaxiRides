```python
rateCodesFilePath = "abfss://de-storage-databrick@databrickstorage1.dfs.core.windows.net/Output/RateCodes.csv"

print(rateCodesFilePath)
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

# Create schema for Rate Codes

rateCodesSchema = (
    StructType
        ([
        StructField("RateCodeId", IntegerType(), True),
        StructField("RateCode", StringType(), True),
        StructField("IsApproved", BooleanType(), True)
    ])
)
```

```python
rateCodesDF = (
    spark
    .read
    .option("header", "true")
    .schema(rateCodesSchema)
    .csv(rateCodesFilePath)  # Ensure the file path exists or handle the error accordingly
)

# Print schema
rateCodesDF.printSchema()
```
