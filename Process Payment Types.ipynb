### A. Define File Path


```python
# MAGIC %md ### A. Define File Path

# COMMAND ----------

# DBTITLE 1,Define file path as variable
# Replace the values: ***ContainerName***, ***DataLakeName*** and file path (if required)

paymentTypesFilePath = "abfss://de-storage-databrick@databrickstorage1.dfs.core.windows.net/Raw/PaymentTypes.json"

print(paymentTypesFilePath)
```

### B. Read File by Applying Schema


```python
# DBTITLE 1,Define schema
from pyspark.sql.types import *
from pyspark.sql.functions import *

# Create schema for Payment Types

paymentTypesSchema = (
    StructType
        ([
        StructField("PaymentTypeID", IntegerType(), True),
        StructField("PaymentType", StringType(), True)
    ])
)

# COMMAND ----------

# DBTITLE 1,Read data by passing schema
# Create DataFrame by applying the schema

paymentTypesDF = (
    spark
    .read

    .schema(paymentTypesSchema)

    .json(paymentTypesFilePath)
)

# Print schema
paymentTypesDF.printSchema()
```

### C. Save Data to Data Lake as Spark (Delta) Table

#### C.1. Create Catalog and Schema in Unity Catalog


```python
(
    paymentTypesDF
    .write

    .mode("overwrite")

    .format("delta")

    .saveAsTable("taxicatalog.rides.paymentTypes")
)
```
