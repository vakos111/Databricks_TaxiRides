### A. Define File Path


```python
taxiZonesFilePath = "abfss://de-storage-databrick@databrickstorage1.dfs.core.windows.net/Raw/TaxiZones.csv"

print(taxiZonesFilePath)
```

```python
# DBTITLE 1,Define schema
from pyspark.sql.types import *
from pyspark.sql.functions import *

# Create schema for Taxi Zones

taxiZonesSchema = (
    StructType
        ([
        StructField("LocationID", IntegerType(), True),
        StructField("Borough", StringType(), True),
        StructField("Zone", StringType(), True),
        StructField("service_zone", StringType(), True)
    ])
)

```

```python
# Create DataFrame by applying the schema

taxiZonesDF = (
    spark
    .read
    .option("header", "true")

    .schema(taxiZonesSchema)

    .csv(taxiZonesFilePath)
)

# Print schema
taxiZonesDF.printSchema()

# COMMAND ----------
```
